```python
#debug後的版本 : 
#自訂函數 1.1：寫一個判斷輸入的正整數是否為質數的函數 is_prime()

def is_prime(x) :
    '''
    這是一個自訂函數，用來判斷輸入的正整數是否為質數
    '''
    try :
        if x <= 0 :
            return "您輸入的是 %s 。不好意思，請輸入正整數!" % x
        elif x == 1 :
            return "1 既不是質數，也不是合數喔!" 
        elif x == 2 :
            return "2 是質數"
        else :
            for a in range(2, x) :
                while x % a == 0 :
                    return "%i 不是質數而是合數喔!" % x 
                    break
            return "%i 是質數" % x     #上次的版本當中，在else底下的return並未縮排，但程式的質數判斷結果並未出錯。在此加以縮排來修正!
    except :
            return "您輸入的是 %s 。輸入的值不是屬於實數就好。還必須是正整數!" % x 


        
        
        
print(is_prime(2))
print(is_prime(3))  
print(is_prime(5))
print(is_prime(7))
print(is_prime(11))
print(is_prime(13))
print(is_prime(17))
print(is_prime(19))
print(is_prime(83))
print(is_prime(89))
print(is_prime(97))
print(is_prime(101))
print(is_prime(-1))
print(is_prime(-1.2)) 
print(is_prime(0))
print(is_prime(0.2))    
print(is_prime(1))    
print(is_prime(2.2)) 
print(is_prime(7.2)) 
print(is_prime(4))
print(is_prime(6))
print(is_prime(8))
print(is_prime(9))      
print(is_prime(10))
print(is_prime(15))     
print(is_prime(21))    
print(is_prime(1000000000000))
print(is_prime(1000000000011))   
```
