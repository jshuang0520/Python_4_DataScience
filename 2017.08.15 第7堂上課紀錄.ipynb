```python
#向 Matplotlib 說嗨（3）
#試著更換 X 軸與 Y 軸的範圍         --> 讓圖不要離邊框太近

import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
sin_x = np.sin(x)

# 作圖
plt.plot(x, sin_x, color = "blue", linewidth = 3, linestyle = "-")

plt.xlim(x.min()*2, x.max()*2)           #更換 X 軸與 Y 軸的範圍
plt.ylim(sin_x.min()*2, sin_x.max()*2)   #因為現在左邊的(x,y)都是負的

# 顯示
plt.show()
```

```python
#向 Matplotlib 說嗨（4）
#試著更換座標軸刻度

import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
sin_x = np.sin(x)

# 作圖
plt.plot(x, sin_x, color = "blue", linewidth = 3, linestyle = "-")
plt.xlim(x.min()*1.2, x.max()*1.2)
plt.ylim(sin_x.min()*1.2, sin_x.max()*1.2)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi]) #更換座標軸刻度，打數學符號當坐標刻度要加$$ --> 見下一個cell
plt.yticks([-1, 0, 1])

# 顯示
plt.show()
```

```python
#向 Matplotlib 說嗨（5）
#試著更換座標軸刻度的顯示標籤

import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
sin_x = np.sin(x)

# 作圖
plt.plot(x, sin_x, color = "blue", linewidth = 3, linestyle = "-")
plt.xlim(x.min()*1.2, x.max()*1.2)
plt.ylim(sin_x.min()*1.2, sin_x.max()*1.2)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ["$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$\pi$"]) 
plt.yticks([-1, 0, 1], ["$-1$", "$0$", "$+1$"])               #更換座標軸刻度的顯示標籤 : 呈現數學符號
# $我要在這裡寫數學符號或方法$    -/pi 是負拍的寫法


# 顯示
plt.show()
```

```python
#向 Matplotlib 說嗨（6）
#試著將圖的外框（spine）移動
#隱藏一個水平、一個垂直
#將剩餘的兩個移動至原點

import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
cos_x, sin_x = np.cos(x), np.sin(x)

# 作圖
plt.plot(x, sin_x, color = "blue", linewidth = 3, linestyle = "-")
plt.xlim(x.min()*1.2, x.max()*1.2)
plt.ylim(sin_x.min()*1.2, sin_x.max()*1.2)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ["$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$\pi$"])
plt.yticks([-1, 0, 1], ["$-1$", "$0$", "$+1$"])


ax = plt.gca() # get current axes    調整座標軸 : 把坐標軸單獨抽出來操作
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))   #移動作標軸
#ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

# 顯示
plt.show()
```

```python
#向 Matplotlib 說嗨（7）
#增加圖例、標題

import numpy as np
import matplotlib.pyplot as plt

# 建立 data
x = np.linspace(-np.pi, np.pi, 100)
sin_x = np.sin(x)

# 作圖
plt.plot(x, sin_x, color = "blue", linewidth = 3, linestyle = "-", label = "sin(x)") #增加圖例就要增加label

plt.xlim(x.min()*1.2, x.max()*1.2)
plt.ylim(sin_x.min()*1.2, sin_x.max()*1.2)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], ["$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$\pi$"])
plt.yticks([-1, 0, 1], ["$-1$", "$0$", "$+1$"])
ax = plt.gca() # get current axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
#ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

#圖例legend是lengend的東西，不是axis的東西，所以要plt.legedn()
plt.legend(loc = 'upper left', frameon = False)  #增加圖例在upper left的位置，frameon = False顯示圖例的時候不要邊框

#增加標題
plt.title("$y = sin(x)$")

# 顯示
plt.show()
```

```python
# plt.bar() 長條圖 : 視覺化 類別變數的相異值、看各類別的數值
import numpy as np
import matplotlib.pyplot as plt

N = 5
ice_cream_sales = np.random.randint(50, 100, size = 5)
ind = np.arange(1, N + 1)
width = 0.5
plt.bar(ind, ice_cream_sales, width, facecolor = 'g')
plt.title('Ice cream sales by flavor')
plt.xticks(ind, ['Matcha', 'Chocolate', 'Vanilla', 'Coco', 'Strawberry'])
for x, y in zip(ind, ice_cream_sales):
    plt.text(x, y + 2, '%i' % y, ha = 'center', va = 'bottom')
plt.ylim(0, ice_cream_sales.max() * 1.2)
plt.yticks([])
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('none')

plt.show()
```

```python
#INSTACODE

import matplotlib.pyplot as plt
import numpy as np

def sigmoid_fun(x):
    return 1 / (1 + np.exp(-x))

def sign_fun(x):
    if x < 0:
        return -1
    elif x == 0:
        return 0
    else:
        return 1

x_arr_s = np.linspace(-10, 10, 100)
sigmoid_x = sigmoid_fun(x_arr_s)
x_arr_1 = np.linspace(-10, -0.3)
x_arr_2 = np.linspace(0.3, 10)
sign_x_1 = list(map(sign_fun, x_arr_1))
sign_x_2 = list(map(sign_fun, x_arr_2))

plt.figure(figsize = (10, 4))
plt.subplot(1, 2, 1)
plt.plot(x_arr_s, sigmoid_x, c = "r", linewidth = 3)
plt.xticks([0]), plt.yticks([0, 0.5, 1])
plt.title("sigmoid function")
plt.subplot(1, 2, 2)
plt.plot(x_arr_1, sign_x_1, c = "b", linewidth = 3)
plt.plot(x_arr_2, sign_x_2, c = "b", linewidth = 3)
plt.scatter([0, 0], [-1, 1], facecolor = "none", edgecolor = "b")
plt.scatter([0], [0], c = "b")
plt.title("sign function")
plt.xticks([0]), plt.yticks([-1, 0, 1])
plt.show()
```
